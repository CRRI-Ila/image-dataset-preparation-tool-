```python
from google.colab import files
import os
import zipfile
import shutil

# Step 1: Upload image files
uploaded = files.upload()

# Step 2: Collect uploaded .jpeg/.jpg files
image_files = [f for f in uploaded.keys() if f.lower().endswith(('.jpeg', '.jpg'))]
image_files.sort()

# Step 3: Rename all to Squirrel_001.jpg, converting .jpeg to .jpg
renamed_files = []
for idx, original_name in enumerate(image_files, start=1):
    new_name = f"Squirrel_{idx:03}.jpg"

    # Only rename if names differ
    if original_name != new_name:
        if original_name.lower().endswith('.jpeg'):
            os.rename(original_name, new_name)
        else:
            # Only copy if new_name is not the same as original_name
            if os.path.abspath(original_name) != os.path.abspath(new_name):
                shutil.copy(original_name, new_name)
    renamed_files.append(new_name)

# Step 4: Create ZIP file
zip_filename = "squirrel_images.zip"
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for f in renamed_files:
        zipf.write(f)

print(f"{len(renamed_files)} files renamed and zipped as '{zip_filename}'")

# Step 5: Download the ZIP
files.download(zip_filename)
```
